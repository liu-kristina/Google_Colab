```python
X_train_fs = X_train.iloc[:, columns]
```

```python
catboost = create_model('catboost', fold=5)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_scores = cross_val_score(catboost, X_train_fs, y_train, cv=5, scoring='f1_macro')
print(f"Cross-Validation Scores: {cv_scores}")
print(f"Mean f1 macro average for CatBoost with feature selection: {cv_scores.mean():.2f}")
```

```python
X_test_fs = X_test[X_train_fs.columns]
```

```python
X_test_fs
```

```python
from hyperopt import hp, fmin, tpe, STATUS_OK, Trials
from catboost import CatBoostClassifier

# Define the search space correctly
space = {
    "depth": hp.choice("depth", [4, 5, 6, 7, 8, 9, 10]),
    "l2_leaf_reg": hp.uniform("l2_leaf_reg", 0.1, 10.0)  # Use continuous range
}

def objective(space):
    try:
        model = CatBoostClassifier(
            depth=space['depth'],
            l2_leaf_reg=space['l2_leaf_reg'],
            iterations=1000,
            verbose=0
        )

        f1 = cross_val_score(model, X_train_fs, y_train, cv=3, scoring='f1').mean()
        return {'loss': -f1, 'status': STATUS_OK}

    except Exception as e:
        print(f"Trial failed due to: {e}")
        return {'loss': float('inf'), 'status': STATUS_OK}

# Track trials
trials = Trials()

# Run Bayesian optimization
best = fmin(fn=objective,
            space=space,
            algo=tpe.suggest,  # Tree Parzen Estimator
            max_evals=20,
            trials=trials)

print("Best parameters:", best)
```
